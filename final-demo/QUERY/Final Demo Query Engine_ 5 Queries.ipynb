### 1) Sentiment: ML query for training data for a super category 

* Query engine is performing sentiment polarity calcuations for the ML features
* prior month
* last three months
* last 12 months

```python
query = "Ans ( nodeid, yr, mn, sales, vol, pm_sales, pm_vol, p3m_sales, p3m_vol, \
p12m_sales, p12m_vol, pm_numreviews, pm_avgrating, p3m_numreviews, \
p3m_avgrating, p12m_numreviews, p12m_avgrating, pm_avgsntp, p3m_avgsntp, \
p12m_avgsntp ) :- \
mlfeatures ( nodeid, yr, mn, sales, vol, pm_sales, pm_vol, p3m_sales, \
p3m_vol, p12m_sales, p12m_vol, pm_numreviews, pm_avgrating, \
p3m_numreviews, p3m_avgrating, p12m_numreviews, p12m_avgrating, \
pm_avgsntp, p3m_avgsntp, p12m_avgsntp ), \
nodeid in (%s),yr=2014,mn=2" %  (super_cat)
df = qengine.get_dataframe(query)
df.head()
```
