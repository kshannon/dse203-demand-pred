## 8) Datalog with S1 predicates (S1.products and S1.sales_agg_yr) 
### 1) dispatched to Postgres wrapper 2) determines natural joins 3) inline with clause to perform additional transformation

```python
# two queries only postgres
#gen = SqlGenerator(sqltype="postgres")
#gen.sqlGenerator(query)
query = "Ans(yr,nodeid,sales,vol,rank_vol) :- \
S1.products(asin,nodeid), \
S1.sales_agg_yr(nodeid,yr,sales,vol,_,_,rank_vol), \
yr=2015, rank_vol<=3"

df = qengine.get_unfolded_dataframe(query)
df.head()
```

## 9) Datalog with two S1, S2 predicates (S1.products,S1.sales_agg_yr,S2.ClassificationInfo)

### 1) each query dispacthed to appropriate wrapper with criteria 2) Combiner does merging 3)  Inline views for Products and ClassificationInfo

```python
query = "Ans(yr,nodeid,classification,level0,sales,vol,rank_vol) :- \
S1.products(asin,nodeid), \
S1.sales_agg_yr(nodeid,yr,sales,vol,_,_,rank_vol),S2.ClassificationInfo(nodeid,classification,level0), \
yr=2015,\
nodeid in (2950,3179,3188,3214,3263,3939,171114,465288,132561011,377550011,6133990011,6343223011,7976022011,8622832011,10806600011)"

df = qengine.get_unfolded_dataframe(query)
df
```

## 10) Passing Conditions to right Sources 

```python
query = "Ans(nodeid) :- \
S1.products(asin,nodeid,yr,mn), \
S1.sales_agg_yr(nodeid,yr,sales,vol,_,_,rank_vol),\
S2.ClassificationInfo(nodeid,classification,level0), \
yr=2015,mn=7"

qengine.break_unfolded_query_with_condition(query)
```
